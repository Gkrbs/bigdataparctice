```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
import igraph as ig
import random
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# read csv

# steam_200K
df = pd.read_csv("../Untitled Folder/steam_200k.csv", names=['UserID', 'Game', 'Action', 'Hours', 'Other'])

# all_games_data
df1 = pd.read_csv("../Untitled Folder/all_games.csv",encoding='utf-8')
df1.columns = ['Name', 'Platform', 'Year', 'summary', 'meta_score', 'user_review']
df1 = df1.drop(['summary'], axis = 1)
all_games_data = df1.query("Year>0")

# vgsales_year_data
df2 = pd.read_csv("../Untitled Folder/vgsales_year.csv",encoding='utf-8')
df2 = df2.drop(['Rank'], axis = 1)
vgsales_year_data = df2.query("Year>0")
vgsales_year_data = df2.drop(['Year'], axis = 1)

# Video_Games_Sales_as_at_22_Dec_2016_data
df3 = pd.read_csv("../Untitled Folder/Video_Games_Sales_as_at_22_Dec_2016.csv",encoding='utf-8')
Video_Games_Sales_as_at_22_Dec_2016_data = df3.query("Year_of_Release>0")

# steam_games_by_year
df4 = pd.read_csv("../Untitled Folder/steam games by year.csv",encoding='utf-8')
steam_games_by_year = df4
```

```python
merge_data1 = pd.merge(all_games_data, vgsales_year_data, on=['Name', 'Platform'], how='left')
merge_data1.fillna(0)
merge_data2 = pd.merge(vgsales_year_data, all_games_data, on=['Name', 'Platform'], how='left')
merge_data2.fillna(0)

merge_data2.head()
```

```python
# all_games_data shows the year by genre until the most recent.
# all_games_data can count the largest amount of games.
temp1 = all_games_data.query("Year<=2020")
temp1 = temp1.drop(['meta_score'],axis =1)
temp1 = temp1.drop(['user_review'],axis =1)

plt.figure(figsize = (5,5))
plt.title('Counting Games', fontsize=15)
plt.xlabel('Year', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.grid(color ='lightgray')
plt.plot(temp1.groupby(['Year']).count(), marker = 'o', color = 'black')

plt.savefig(f'./Counting Games by Year.png')

#게임 출시는 계속 늘어나고 있다.
```

```python
#'Name', 'Platform', 'Year', 'meta_score', 'user_review', 'Genre',
#       'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales',
#       'Global_Sales'

temp = merge_data1
temp = temp.drop(['Name', 'Year', 'meta_score', 'user_review',
                  'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales',
                  'Other_Sales'],axis =1)
temp.rename(columns = {'Global_Sales':''}, inplace = True)
temp = temp.groupby(['Platform','Genre']).count().unstack(level=0)

sns.set(rc = {'figure.figsize':(30,15)}, font_scale=2)
heatmaptemp = sns.heatmap(temp, linewidths=0,
                         linecolor = 'white',vmin = 0, 
                          vmax = 100, cmap = 'Reds', fmt ='.0f')
heatmaptemp.set(xlabel="Platform", ylabel="Genre")

plt.savefig(f'./Genre Platform Association.png')
```
